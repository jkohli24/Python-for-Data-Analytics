We are using tweetering.py to get the data for a specifc word and wil clean the data using regular expressions . 
I am converting the file into the list and basically comparing it with the stop words list , positive word list and negative word list . 

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import re #regular expression
```

```python
#Reading the FILE
file =open("trump.txt")
tweet_file = file.readline() # reading the file 
tweet_file=tweet_file.lower() # converting the file to lower case 
tweet_file=tweet_file.strip() # removing the whitespaces from the file 
tweet_file=tweet_file.replace("@","").replace(":","").replace(")","").replace(".","").replace("!","").replace(",","").\
replace("?","").replace("#","").replace("\\","").replace("%","").replace("\n","").replace("_","").\
replace("|","").replace("n\''","").replace("-","").replace('(',"").replace('rt',"").replace('" "'," ").replace('/',"").replace('\/',"").\
replace('//t.c', "").replace('\\x[0-9a-f][0-9a-f]', "")
# I used replace function to remove all the gibberish from the tweets and cleaning the data to make it ready for analysis
```

```python
Twitter= [] # creating the list 
Twitter= tweet_file.split()# converting the file to list 
```

```python
Twitter    
```

```python
stopWords = [] #creating the list 
with open('stopwords.txt', 'r') as f: #reading the stopwords data 
    for word in f:
        word = word.split('\n') #spliting the data so that each line has only one word 
        stopWords.append(word[0]) #Using the append function to add the words into the list .
```

```python
positive = []#reading the positive data 
with open('positive.txt', 'r') as f:
    for word in f:
        word = word.split('\n')#spliting the data so that each line has only one word 
        positive.append(word[0])#Using the append function to add the words into the list .
```

```python
negative = [] #reading the negative data 
with open('negative.txt', 'r') as f:
    for word in f:
        word = word.split('\n') #spliting the data so that each line has only one word 
        negative.append(word[0])#Using the append function to add the words into the list 
```

```python
 (stopcount, positivecount,negativecount,others,sentiment) = (0,0,0,0,0) # setting counter = 0 So that it doesnt return garbage value 

    

## Loop to check the words present in all other files and increasing the counter
for word in Twitter:
    if  word in stopWords:
        stopcount +=1
    elif word in negative: 
        negativecount  +=1
        sentiment -= 1
    elif word in positive:
        positivecount +=1
        sentiment += 1
    else :
            others +=1        

print("Sentiment value: " + str(sentiment))
print("Number of stop count:" + str(stopcount))
print("Number of positive count:" + str(positivecount))
print("Number of negative count:" + str(negativecount))
print("Number of others count:" + str(others))

```

```python
totalcount = len(Twitter)
print("Ratio of positive words to total:" +str(round(positivecount/totalcount,2)))
print("Ratio of negative words to total:" +str(round(negativecount/totalcount,2)))
print("Ratio of stop words to total:" +str(round(stopcount/totalcount,2)))

print("Ratio of positive words to negative:" +str(round(positivecount/negativecount,2)))
```

```python
if negativecount>positivecount:
    print("Sentiment of given word is negative")
elif positivecount>negativecount:
    print("Sentiment of given word is positive")
else:
     print("Sentiment of given word is neutral")
```

# <font color='green'>Trump word has a Positive sentiment accoring to the tweets collected </font>
